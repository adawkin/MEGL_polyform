```python
import numpy as np
import random as rm
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
```

```python
# list of possible 3-minos, t = 1 transitions, and their probabilities

mino_3 = ["A", "B", "C", "D", "E", "F"]

transition = [["AA", "AB", "AC", "AD", "AE", "AF"], 
               ["BA", "BB", "BC", "BD", "BE", "BF"], 
               ["CA", "CB", "CC", "CD", "CE", "CF"], 
               ["DA", "DB", "DC", "DD", "DE", "DF"], 
               ["EA", "EB", "EC", "ED", "EE", "EF"], 
               ["FA", "FB", "FC", "FD", "FE", "FF"]]

transition_prob = [[5/9, 0, 1/9, 1/9, 1/9, 1/9],
                   [0, 5/9, 1/9, 1/9, 1/9, 1/9], 
                   [1/9, 1/9, 7/18, 1/9, 1/9, 1/6],
                   [1/9, 1/9, 1/9, 7/18, 1/6, 1/9],
                   [1/9, 1/9, 1/9, 1/6, 7/18, 1/9],
                   [1/9, 1/9, 1/6, 1/9, 1/9, 7/18]]
```

```python
# markov chain, given a starting 3-mino, and a number of runs, return a potential sequence

def mino_chance(start, runs):
    
    mino_current = start
    
    mino_list = [mino_current]
    
    i = 0
    
    prob = 1
    
    while i != runs:
        if mino_current == "A":
            change = np.random.choice(transition[0],replace=True,p=transition_prob[0])
            if change == "AA":
                prob *= 5/9
                mino_current = "A"
                mino_list.append("A")
            if change == "AB":
                prob *= 0
                mino_current = "B"
                mino_list.append("B")
            if change == "AC":
                prob *= 1/9
                mino_current = "C"
                mino_list.append("C")
            if change == "AD":
                prob *= 1/9
                mino_current = "D"
                mino_list.append("D")
            if change == "AE":
                prob *= 1/9
                mino_current = "E"
                mino_list.append("E")
            if change == "AF":
                prob *= 1/9
                mino_current = "F"
                mino_list.append("F")
                
        elif mino_current == "B":
            change = np.random.choice(transition[1],replace=True,p=transition_prob[1])
            if change == "BA":
                prob *= 0
                mino_current = "A"
                mino_list.append("A")
            if change == "BB":
                prob *= 5/9
                mino_current = "B"
                mino_list.append("B")
            if change == "BC":
                prob *= 1/9
                mino_current = "C"
                mino_list.append("C")
            if change == "BD":
                prob *= 1/9
                mino_current = "D"
                mino_list.append("D")
            if change == "BE":
                prob *= 1/9
                mino_current = "E"
                mino_list.append("E")
            if change == "BF":
                prob *= 1/9
                mino_current = "F"
                mino_list.append("F")

        elif mino_current == "C":
            change = np.random.choice(transition[2],replace=True,p=transition_prob[2])
            if change == "CA":
                prob *= 1/9
                mino_current = "A"
                mino_list.append("A")
            if change == "CB":
                prob *= 1/9
                mino_current = "B"
                mino_list.append("B")
            if change == "CC":
                prob *= 7/18
                mino_current = "C"
                mino_list.append("C")
            if change == "CD":
                prob *= 1/9
                mino_current = "D"
                mino_list.append("D")
            if change == "CE":
                prob *= 1/9
                mino_current = "E"
                mino_list.append("E")
            if change == "CF":
                prob *= 1/6
                mino_current = "F"
                mino_list.append("F")
            
        elif mino_current == "D":
            change = np.random.choice(transition[3],replace=True,p=transition_prob[3])
            if change == "DA":
                prob *= 1/9
                mino_current = "A"
                mino_list.append("A")
            if change == "DB":
                prob *= 1/9
                mino_current = "B"
                mino_list.append("B")
            if change == "DC":
                prob *= 1/9
                mino_current = "C"
                mino_list.append("C")
            if change == "DD":
                prob *= 7/18
                mino_current = "D"
                mino_list.append("D")
            if change == "DE":
                prob *= 1/6
                mino_current = "E"
                mino_list.append("E")
            if change == "DF":
                prob *= 1/9
                mino_current = "F"
                mino_list.append("F")
            
        elif mino_current == "E":
            change = np.random.choice(transition[4],replace=True,p=transition_prob[4])
            if change == "EA":
                prob *= 1/9
                mino_current = "A"
                mino_list.append("A")
            if change == "EB":
                prob *= 1/9
                mino_current = "B"
                mino_list.append("B")
            if change == "EC":
                prob *= 1/9
                mino_current = "C"
                mino_list.append("C")
            if change == "ED":
                prob *= 1/6
                mino_current = "D"
                mino_list.append("D")
            if change == "EE":
                prob *= 7/18
                mino_current = "E"
                mino_list.append("E")
            if change == "EF":
                prob *= 1/9
                mino_current = "F"
                mino_list.append("F")
            
            
        elif mino_current == "F":
            change = np.random.choice(transition[5],replace=True,p=transition_prob[5])
            if change == "FA":
                prob *= 1/9
                mino_current = "A"
                mino_list.append("A")
            if change == "FB":
                prob *= 1/9
                mino_current = "B"
                mino_list.append("B")
            if change == "FC":
                prob *= 1/6
                mino_current = "C"
                mino_list.append("C")
            if change == "FD":
                prob *= 1/9
                mino_current = "D"
                mino_list.append("D")
            if change == "FE":
                prob *= 1/9
                mino_current = "E"
                mino_list.append("E")
            if change == "FF":
                prob *= 7/18
                mino_current = "F"
                mino_list.append("F")

        i += 1
        
#     print("The sequence of 3 - minos is: " + str(mino_list))
#     print("The last mino is: " + str(mino_current))
#     print("Probability of this sequence is: " + str(prob))
    
    return mino_list
```

```python
# using mino_chance, simulates the probability of starting at a given mino and ending at a given mino with a certain number of runs
# n is the number of simulations

def mino_sim(start, end, runs, n):
    list_mino_list = []
    count = 0
    for i in range(1, n):
        list_mino_list.append(mino_chance(start, runs))
        
        
    for a_list in list_mino_list:
        if a_list[runs] == end:
            count += 1
            
    percent = (count/n)
    
    print("The probability of starting at: " + start + " and ending at " + end + " is " +str(percent))
```

```python
mino_sim("A", "A", 2, 10000)
mino_sim("A", "B", 2, 10000)
mino_sim("A", "C", 2, 10000)
mino_sim("A", "D", 2, 10000)
mino_sim("A", "E", 2, 10000)
mino_sim("A", "F", 2, 10000)
```

```python
# creates an empty list for the purpose of filling empty values in subsequent dictionaries
mino_count = [0] * len(mino_3)
```

```python
# potential sequence and histogram of such sequence given 1 run

hist_1 = mino_chance("A", 1)
print("After 1 step, the last mino is " + str(hist_1[-1]) + ".")

hist_1_dict = dict(zip(mino_3, mino_count))
hist_1_dict.update(Counter(hist_1))

hist_1_df = pd.DataFrame.from_dict(hist_1_dict, orient = "index")
hist_1_df.plot(kind = "bar", legend = False)
```
